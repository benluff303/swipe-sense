```python
keywords = [
    ["fun", "cosmopolitan", "nightlife", "busy"],
    ["relaxing", "beach", "scenic", "romantic"],
    ["peaceful", "mountains", "quiet", "breathtaking"],
    ["educational", "ancient", "ruins", "unforgettable"],
    ["charming", "coastal", "laid-back", "ocean"],
    ["wild", "exciting", "party", "locals"],
    ["calm", "lake", "beautiful", "tranquil"],
    ["city", "vibe", "action", "hotel"],
    ["spa", "treatments", "rejuvenating", "relaxing"],
    ["romantic", "sunset", "cocktails", "rooftop"],
    ["market", "vibrant", "crafts", "unique"],
    ["gardens", "lush", "scenic", "peaceful"],
    ["kayaking", "river", "fun", "calming"],
    ["special", "staff", "stay", "relaxing"],
    ["exploring", "local", "adventure", "experience"]
]
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import re
from collections import defaultdict
```

```python
reviews = df["review_text"].tolist()
```

```python
reviews = [str(r) if r is not None else "" for r in reviews]
```

```python
df = df.reset_index(drop=True)

# Encode reviews
review_embeddings = model.encode(reviews, convert_to_tensor=True)
```

```python
# Encode queries
keywords = [" ".join(k) for k in keywords]
keyword_embeddings = model.encode(keywords, convert_to_tensor=True)

# Collect results
results = []

for i, query_emb in enumerate(keyword_embeddings):
    similarities = util.cos_sim(query_emb, review_embeddings)[0]

    # Aggregate similarities by city
    city_sims = defaultdict(list)
    for idx, sim in enumerate(similarities):
        city = df.iloc[idx]['location']
        city_sims[city].append((sim.item(), idx))

    # Compute average similarity per city
    city_avg_scores = []
    for city, sims in city_sims.items():
        avg_score = sum([s[0] for s in sims]) / len(sims)
        city_avg_scores.append((city, avg_score, sims))

    # Sort cities by average similarity
    city_avg_scores_sorted = sorted(city_avg_scores, key=lambda x: x[1], reverse=True)[:5]

    # For each top city, pick top review
    for city, avg_score, sims in city_avg_scores_sorted:
        sims_sorted = sorted(sims, key=lambda x: x[0], reverse=True)[:1]  # top review
        for score, idx in sims_sorted:
            row = df.iloc[idx]
            results.append({
                'keywords': keywords[i],
                'location': city,
                'place_name': row['place_name'],
                'review': row['review_text'],
                'review_score': score,
                'city_score': avg_score
            })

# Create dataframe
top_cities_df = pd.DataFrame(results)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")  # efficient, 384-dim
embeddings = model.encode(reviews, show_progress_bar=True, batch_size=64)
```

```python
import hdbscan

clusterer = hdbscan.HDBSCAN(min_cluster_size=30, metric='euclidean')
labels = clusterer.fit_predict(embeddings)
```

```python
from keybert import KeyBERT
kw_model = KeyBERT(model)

cluster_keywords = {}
for label in set(labels):
    cluster_texts = [reviews[i] for i in range(len(reviews)) if labels[i] == label]
    if not cluster_texts: continue
    docs_joined = " ".join(cluster_texts)
    keywords = kw_model.extract_keywords(
        docs_joined, keyphrase_ngram_range=(1, 2),
        stop_words="english", top_n=10
    )
    cluster_keywords[label] = keywords

print(cluster_keywords[0])
```
