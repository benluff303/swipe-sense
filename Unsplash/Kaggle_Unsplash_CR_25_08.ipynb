```python
import ast

def get_review_text(x):
    if isinstance(x, dict):
        return x.get('text', '')
    elif isinstance(x, str):
        try:
            d = ast.literal_eval(x)
            if isinstance(d, dict) and 'text' in d:
                return d['text']
            else:
                return x
        except Exception:
            return x
    else:
        return ''
```

```python
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from sentence_transformers import SentenceTransformer, InputExample

reviews = revs['reviews']

#faketags
tags = [
    ["fun", "cosmopolitan", "nightlife", "busy"],
    ["relaxing", "beach", "scenic", "romantic"],
    ["peaceful", "mountains", "quiet", "breathtaking"],
    ["educational", "ancient", "ruins", "unforgettable"],
    ["charming", "coastal", "laid-back", "ocean"],
    ["wild", "exciting", "party", "locals"],
    ["calm", "lake", "beautiful", "tranquil"],
    ["city", "vibe", "action", "hotel"],
    ["spa", "treatments", "rejuvenating", "relaxing"],
    ["romantic", "sunset", "cocktails", "rooftop"],
    ["market", "vibrant", "crafts", "unique"],
    ["gardens", "lush", "scenic", "peaceful"],
    ["kayaking", "river", "fun", "calming"],
    ["special", "staff", "stay", "relaxing"],
    ["exploring", "local", "adventure", "experience"]
]

all_tags = sorted(list(set(tag for taglist in tags for tag in taglist)))
tag2idx = {tag: i for i, tag in enumerate(all_tags)}
num_labels = len(all_tags)
```

```python
class ReviewDataset(Dataset):
    def __init__(self, reviews, tags, tag2idx):
        # reviews and tags should be lists, not pandas Series
        self.reviews = list(reviews)
        self.tags = list(tags)
        self.tag2idx = tag2idx

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, idx):
        text = self.reviews[idx]
        tag_vec = [0] * len(self.tag2idx)
        # Handle missing tags by assigning empty list if needed
        if idx < len(self.tags) and self.tags[idx]:
            for tag in self.tags[idx]:
                if tag in self.tag2idx:
                    tag_vec[self.tag2idx[tag]] = 1
        return text, torch.tensor(tag_vec, dtype=torch.float)

# Instantiate SBERT and dataset
model = SentenceTransformer('all-MiniLM-L6-v2')
dataset = ReviewDataset(reviews, tags, tag2idx)

def custom_collate(batch):
    texts = [item[0] for item in batch]      # list of strings
    targets = torch.stack([item[1] for item in batch])  # tensor
    return texts, targets

dataloader = DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=custom_collate)
```

```python
from transformers import AutoTokenizer

class SBERTMultiLabelClassifier(nn.Module):
    def __init__(self, sbert_model, num_labels):
        super().__init__()
        self.sbert_model = sbert_model._first_module().auto_model
        # This works for all official SentenceTransformer models:
        model_name = sbert_model._first_module().auto_model.config._name_or_path
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.classifier = nn.Linear(self.sbert_model.config.hidden_size, num_labels)

    def forward(self, texts):
        encoding = self.tokenizer(
            texts,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        )
        encoding = {k: v.to(next(self.parameters()).device) for k, v in encoding.items()}
        outputs = self.sbert_model(**encoding)
        embeddings = outputs.last_hidden_state[:, 0, :]
        logits = self.classifier(embeddings)
        return logits

clf_model = SBERTMultiLabelClassifier(model, num_labels)
optimizer = torch.optim.Adam(clf_model.parameters(), lr=2e-5)
criterion = nn.BCEWithLogitsLoss()
```

```python
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

clf_model = SBERTMultiLabelClassifier(model, num_labels).to(device)
optimizer = torch.optim.Adam(clf_model.parameters(), lr=2e-5)

for epoch in range(3):
    clf_model.train()
    for texts, targets in dataloader:
        targets = targets.to(device)
        logits = clf_model(texts)   # classifier's forward moves encoding to device
        loss = criterion(logits, targets)
        loss.backward()
        optimizer.step()
```

```python
# Prediction
clf_model.eval()
sample_review = "The nightlife was fun and the sunset was romantic."
with torch.no_grad():
    logits = clf_model([sample_review])
    probs = torch.sigmoid(logits).cpu().numpy()[0]
    pred_tags = [tag for tag, p in zip(all_tags, probs) if p > 0.35]
    print("Predicted tags:", pred_tags)
```
