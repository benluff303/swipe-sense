```python
reviews = df["review_text"].tolist()
```

```python
reviews = [str(r) if r is not None else "" for r in reviews]
```

```python
df = df.reset_index(drop=True)

# Encode reviews
review_embeddings = model.encode(reviews, convert_to_tensor=True)
```

```python
# Encode queries
keywords = [" ".join(k) for k in keywords]
keyword_embeddings = model.encode(keywords, convert_to_tensor=True)

# Collect results
results = []

for i, query_emb in enumerate(keyword_embeddings):
    similarities = util.cos_sim(query_emb, review_embeddings)[0]

    # Aggregate similarities by city
    city_sims = defaultdict(list)
    for idx, sim in enumerate(similarities):
        city = df.iloc[idx]['location']
        city_sims[city].append((sim.item(), idx))

    # Compute average similarity per city
    city_avg_scores = []
    for city, sims in city_sims.items():
        avg_score = sum([s[0] for s in sims]) / len(sims)
        city_avg_scores.append((city, avg_score, sims))

    # Sort cities by average similarity
    city_avg_scores_sorted = sorted(city_avg_scores, key=lambda x: x[1], reverse=True)[:5]

    # For each top city, pick top review
    for city, avg_score, sims in city_avg_scores_sorted:
        sims_sorted = sorted(sims, key=lambda x: x[0], reverse=True)[:1]  # top review
        for score, idx in sims_sorted:
            row = df.iloc[idx]
            results.append({
                'keywords': keywords[i],
                'location': city,
                'place_name': row['place_name'],
                'review': row['review_text'],
                'review_score': score,
                'city_score': avg_score
            })

# Create dataframe
top_cities_df = pd.DataFrame(results)
```

```python
from sentence_transformers import SentenceTransformer, util
import re
import pandas as pd
import numpy as np
from collections import defaultdict
import torch

def top_locations(df, keywords):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    keyword_strings = [" ".join(k) if isinstance(k, list) else str(k) for k in keywords]
    list_reviews = df['review_text'].tolist()
    reviews = [str(r) if r is not None else "" for r in list_reviews]
    review_embeddings = model.encode(reviews)
    keyword_embeddings = model.encode(keyword_strings)
    np.save('review_embeddings.npy', review_embeddings)
    np.save('keyword_embeddings.npy', keyword_embeddings)
    review_embeddings_np = np.load("review_embeddings.npy")
    keyword_embeddings_np = np.load("keyword_embeddings.npy")

    # Convert to tensors
    review_embeddings = torch.tensor(review_embeddings_np)
    keyword_embeddings = torch.tensor(keyword_embeddings_np)

    # Collect results
    results = []

    # Compute the cosine similarity for keyword group embeddings to all review embeddings.
    for i, query_emb in enumerate(keyword_embeddings):
        similarities = util.cos_sim(query_emb, review_embeddings)[0]

        # Aggregate similarities by location
        location_sims = defaultdict(list)
        for idx, sim in enumerate(similarities):
            location = df.iloc[idx]['location']
            location_sims[location].append((sim.item(), idx))

        # Compute average similarity per location
        location_avg_scores = []
        for location, sims in location_sims.items():
            avg_score = sum([s[0] for s in sims]) / len(sims)
            location_avg_scores.append((location, avg_score, sims))

        # Sort locations by average similarity
        location_avg_scores_sorted = sorted(location_avg_scores, key=lambda x: x[1], reverse=True)[:5]

        # For each top location, pick top review
        for location, avg_score, sims in location_avg_scores_sorted:
                results.append({
                    'keywords': keyword_strings[i],
                    'location': location,
                    'location_score': avg_score
                })

    return pd.DataFrame(results)
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

def semantic_word_strength(df_reviews, df_words):
    model = SentenceTransformer("all-MiniLM-L6-v2")

    # Candidate words
    candidate_words = df_words['Vibes'].tolist() + \
                      df_words['Activities'].tolist() + \
                      df_words['Destination Features'].tolist()

    # Encode reviews + words
    review_embeddings = model.encode([str(r) for r in df_reviews['review_text']], convert_to_tensor=True)
    word_embeddings = model.encode(candidate_words, convert_to_tensor=True)

    # For each word, compute average similarity with all reviews
    scores = []
    for i, word in enumerate(candidate_words):
        sims = util.cos_sim(word_embeddings[i], review_embeddings)[0]
        avg_score = torch.mean(sims).item()
        scores.append((word, avg_score))

    # Return sorted dataframe
    return pd.DataFrame(scores, columns=["word", "similarity_score"]).sort_values(by="similarity_score", ascending=False)
```
