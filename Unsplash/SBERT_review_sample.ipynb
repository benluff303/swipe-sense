```python
keywords = [
    ["fun", "cosmopolitan", "nightlife", "busy"],
    ["relaxing", "beach", "scenic", "romantic"],
    ["peaceful", "mountains", "quiet", "breathtaking"],
    ["educational", "ancient", "ruins", "unforgettable"],
    ["charming", "coastal", "laid-back", "ocean"],
    ["wild", "exciting", "party", "locals"],
    ["calm", "lake", "beautiful", "tranquil"],
    ["city", "vibe", "action", "hotel"],
    ["spa", "treatments", "rejuvenating", "relaxing"],
    ["romantic", "sunset", "cocktails", "rooftop"],
    ["market", "vibrant", "crafts", "unique"],
    ["gardens", "lush", "scenic", "peaceful"],
    ["kayaking", "river", "fun", "calming"],
    ["special", "staff", "stay", "relaxing"],
    ["exploring", "local", "adventure", "experience"]
]
```

```python
import pandas as pd
revs = pd.read_csv("sample_location_activity_reviews.csv")
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
revs['rating'].value_counts()
```

```python
revs = revs[revs['rating'] >= 4]
```

```python
revs
```

```python
import re

# Extract everything after 'in' (case-insensitive) as the city
revs['city'] = revs['query'].apply(
    lambda x: re.split(r'\bin\b', str(x), flags=re.IGNORECASE)[-1].strip()
)
revs
```

```python
revs.columns
```

```python
revs = revs.reset_index(drop=True)
reviews = revs['reviews']

# Encode reviews
review_embeddings = model.encode(reviews)

# Encode queries
query_strings = [" ".join(k) for k in keywords]
query_embeddings = model.encode(query_strings)

results = []

for i, query_emb in enumerate(query_embeddings):
    similarities = util.cos_sim(query_emb, review_embeddings)[0]
    sorted_idx = similarities.argsort(descending=True).cpu().numpy().tolist()

    seen_cities = set()
    top_results = []

    for idx in sorted_idx:
        row = revs.iloc[idx]
        city = row['city']
        if city not in seen_cities:   # enforce unique city
            seen_cities.add(city)
            top_results.append((row, similarities[idx]))
        if len(top_results) == 5:
            break

    for row, score in top_results:
        results.append({
            'query_string': query_strings[i],
            'city': row['city'],
            'place_name': row['place_name'],
            'review': row['reviews'],
            'score': score.item()  # convert tensor to float
        })

# Create dataframe
top_reviews_df = pd.DataFrame(results)
```

```python
import pandas as pd
import re
from collections import defaultdict

revs = revs.reset_index(drop=True)
reviews = revs['reviews']

# Encode reviews
review_embeddings = model.encode(reviews)

# Encode queries
query_strings = [" ".join(k) for k in keywords]
query_embeddings = model.encode(query_strings)

# Collect results
results = []

for i, query_emb in enumerate(query_embeddings):
    similarities = util.cos_sim(query_emb, review_embeddings)[0]

    # Aggregate similarities by city
    city_sims = defaultdict(list)
    for idx, sim in enumerate(similarities):
        city = revs.iloc[idx]['city']
        city_sims[city].append((sim.item(), idx))

    # Compute average similarity per city
    city_avg_scores = []
    for city, sims in city_sims.items():
        avg_score = sum([s[0] for s in sims]) / len(sims)
        city_avg_scores.append((city, avg_score, sims))

    # Sort cities by average similarity
    city_avg_scores_sorted = sorted(city_avg_scores, key=lambda x: x[1], reverse=True)[:5]

    # For each top city, pick top review
    for city, avg_score, sims in city_avg_scores_sorted:
        sims_sorted = sorted(sims, key=lambda x: x[0], reverse=True)[:1]  # top review
        for score, idx in sims_sorted:
            row = revs.iloc[idx]
            results.append({
                'query_string': query_strings[i],
                'city': city,
                'place_name': row['place_name'],
                'review': row['reviews'],
                'review_score': score,
                'city_score': avg_score
            })

# Create dataframe
top_cities_df = pd.DataFrame(results)
```
